## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Write your code here
data = pd.read_csv("../data/company_sales_data.csv")
data.head()
df = data.copy()
```

```python
X = data['month_number']
Y = data['total_profit']
# plot
plt.plot(X, Y)
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.title('Company profit per month')
plt.xticks(X)
plt.ylim(100000, 500000)
```

## Exercise 2
**Get total profit of all months and show line plot with the following Style properties**
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number

```python
# write your code here
plt.plot(X, data['facecream'], color = 'blue', marker = 'o')
plt.plot(X, data['facewash'], color = 'orange', marker = 'o')
plt.plot(X, data['toothpaste'], color = 'g', marker = 'o')
plt.plot(X, data['bathingsoap'], color = 'r', marker = 'o')
plt.plot(X, data['shampoo'], color = 'purple', marker = 'o')
plt.plot(X, data['moisturizer'], color = 'brown', marker = 'o')
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('Sales data')
plt.xticks(X)
plt.ylim(1000, 18000)
```

## Exercise 4
**Read toothpaste sales data of each month and show it using a scatter plot**
Also, add a grid in the plot. gridline style should .

**Expected output**

<img src="../data/Tooth_paste_Sales_data.png" alt="sales_data" style="width: 700px;"/>


```python
# write your code here
toothpaste = data['toothpaste']
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.scatter(X, toothpaste, label = 'Toothpaste sales data')
ax.set_xlabel("Month Number")
ax.set_ylabel("Number of units sold")
ax.set_title("Toothpaste sales data")
plt.legend(loc = 'upper left')
plt.grid(linestyle='--')
plt.xticks(X)
plt.ylim(4500,8000)
plt.show()
```

## Exercise 5
**Read face cream and facewash product sales data and show it using the bar chart**
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

**Expected output**

<img src="../data/Facewash_and_facecream_sales_data.png" alt="sales_data" style="width: 700px;"/>


```python
# write your code here
fc = data['facecream']
fw = data['facewash']
plt.figure(figsize = (8,8))
plt.bar(X - 0.2, fc, 0.4, label = 'Facecream')
plt.bar(X + 0.2, fw, 0.4, label = 'Facewash')
plt.xlabel('Month Number')
plt.ylabel('Series units in number')
plt.title('Facecream and Facewash sales data')
plt.xticks(X)
plt.ylim(0,3500)
plt.grid(linestyle = '--')
plt.legend(loc = 'upper left')
plt.show()
```

**Calculate total sale data for last year for each product and show it using a Pie chart**
